File Pointer

```python


with open("./sample_data/test2.txt", "r") as file:  # ✅ "r" = read mode
    print(file.tell())  # Current cursor position
    print(file.read())   # Read the content
    print(file.tell())   # Cursor after reading
```

```python
with open("./sample_data/sample.txt", "r") as file:  # ✅ "r" = read mode
    print(file.tell())  # Current cursor position
    print(file.read())   # Read the content
    print(file.tell())   # Cursor after reading
    print(file.read()) # last index theke pora shuru korsi
```

```python
with open("./sample_data/sample.txt", "r") as file:  # ✅ "r" = read mode
    print(file.tell())  # Current cursor position
    print(file.read(5))
    print(file.tell())

    print(file.read())
    print(file.tell())
```

```python
with open("./sample_data/sample.txt", "r") as file:  # ✅ "r" = read mode
    print(file.tell())  # Current cursor position

    file.seek(5)
    print(file.tell())
```

```python
with open("./sample_data/sample.txt", "r") as file:  # ✅ "r" = read mode
    print(file.tell())  # Current cursor position

    print(file.read(5))

    print(file.tell())

    file.seek(0)
    print(file.tell())
    print(file.read(5))

```
